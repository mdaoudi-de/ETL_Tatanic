```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# afficher toutes les colonnes (ne pas tronquer)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
pd.set_option("display.width", None)
```

```python
# Chargement des données

csv_path = Path("../data/raw/titanic.csv")  # adapte si besoin
df = pd.read_csv(csv_path)
df.head(15)
```

```python
# Aperçu des données
print(df.dtypes)
```

```python
# Selection des colonnes utiles
df = df[["Passengerid", "Age", "Fare", "Sex", "Pclass", "sibsp", "Parch", "Pclass","Embarked","2urvived"]]
print(df.dtypes)
```

```python
# Aperçu des données
df.head(15)
```

```python
# Aperçu info & mémoire
df.info()
mem_mb = df.memory_usage(deep=True).sum() / 1e6
print(f"\nMémoire ~ {mem_mb:.2f} MB")
```

```python
# Suppression des lignes avec des valeurs manquantes dans la colonne "Embarked"
df = df.dropna(subset=["Embarked"])
```

```python
# Suppression des doublons
dup_count = df.duplicated().sum()
print("Lignes dupliquées:", dup_count)
if dup_count:
    display(df[df.duplicated()].head())
```

```python
df.info()
mem_mb = df.memory_usage(deep=True).sum() / 1e6
print(f"\nMémoire ~ {mem_mb:.2f} MB")
```
